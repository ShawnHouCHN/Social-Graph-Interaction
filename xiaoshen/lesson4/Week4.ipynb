    ... some text [[Aristotle]] some more text ...

```python
import io
import re
from __future__ import print_function 
#open files
aes_file = io.open('aestheticians.txt','r',encoding='utf-8')
epi_file = io.open('epistemologists.txt','r',encoding='utf-8')
eth_file = io.open('ethicists.txt','r',encoding='utf-8')
log_file = io.open('logicians.txt','r',encoding='utf-8')
met_file = io.open('metaphysicians.txt','r',encoding='utf-8')
sap_file = io.open('social and political philosophers.txt','r',encoding='utf-8')

aes_phys = re.findall(r'\*\s*([\w\.\s]*)(\[*)([\w\s\.-]+)([\w\s\.|()-,]*)(\]\])', aes_file.read(),re.UNICODE)
epi_phys = re.findall(r'\*\s*([\w\.\s]*)(\[*)([\w\s\.-]+)([\w\s\.|()-,]*)(\]\])', epi_file.read(),re.UNICODE)
eth_phys = re.findall(r'\*\s*([\w\.\s]*)(\[*)([\w\s\.-]+)([\w\s\.|()-,]*)(\]\])', eth_file.read(),re.UNICODE)
log_phys = re.findall(r'\*\s*([\w\.\s]*)(\[*)([\w\s\.-]+)([\w\s\.|()-,]*)(\]\])', log_file.read(),re.UNICODE) #log
met_phys = re.findall(r'\*\s*([\w\.\s]*)(\[*)([\w\s\.-]+)([\w\s\.|()-,]*)(\]\])', met_file.read(),re.UNICODE) #met
sap_phys = re.findall(r'\*\s*([\w\.\s]*)(\[*)([\w\s\.-]+)([\w\s\.|()-,]*)(\]\])', sap_file.read(),re.UNICODE)

comb_list=[]
comb_aes_list=[]
comb_epi_list=[]
comb_eth_list=[]
comb_log_list=[]
comb_met_list=[]
comb_sap_list=[]
#print len(aes_phys)
for aes_phy in aes_phys:  #print each phyloer of the list 
    comb_aes_list.append(aes_phy[2])
for epi_phy in epi_phys:  #print each phyloer of the list 
    comb_epi_list.append(epi_phy[2])
for eth_phy in eth_phys:  #print each phyloer of the list 
    comb_eth_list.append(eth_phy[2])
for log_phy in log_phys:  #print each phyloer of the list 
    comb_log_list.append(log_phy[2])
for met_phy in met_phys:  #print each phyloer of the list 
    comb_met_list.append(met_phy[2])
for sap_phy in sap_phys:  #print each phyloer of the list 
    comb_sap_list.append(sap_phy[2])
comb_list=set(comb_aes_list+comb_epi_list + comb_met_list + comb_eth_list + comb_log_list + comb_sap_list)
#epi_phys = re.findall(r'\*([\w\.\s]*)(\[\[)([\w\s]+)([\w\s|()]*)(\]\])', epi_file.read(),re.UNICODE)
for log_phy in log_phys:
    print (log_phy[2])
print (len(comb_list))

#size of each list
print("size of aes",len(aes_phys))
print("size of epi",len(epi_phys))
print("size of eth",len(eth_phys))
print("size of log",len(log_phys))
print("size of met",len(met_phys))
print("size of sap",len(sap_phys))
```

```python
import urllib
import urllib2
baseurl="https://en.wikipedia.org/w/api.php?"
action="action=query"
content="prop=revisions&rvprop=content"
dataformat="format=json"
for phy in comb_list:
    titles = "titles="+urllib.quote(phy.encode('utf-8'))
    query = "%s&%s&%s&%s&%s" % (baseurl,action,titles,content,dataformat)
    #print (query)
    page=urllib2.urlopen(query).read()
    with open('philosophers/'+titles+'.json','wb') as output:
        output.write(page)
```

# Part B: Building the networks

Now, we're going to build a NetworkX directed graph of the links between pages. For each philosopher page, we'll find all the links to other characters, and for each link add an edge in the graph. 
> 
> _Exercise_: Build the philosopher network. The **combined list** from above provides the network nodes. We will use info from the pages to define links. Every time philosopher A's page links to philosopher B's page, we create a directed link from A to B. There are many ways to do this, but below, I've tried to break it down into natural steps.
> 
> * Use a regular expression to extract all outgoing links from each of the pages you downloaded above. There are many ways to iterate over pages - feel free to choose the one that matches the way you've chosen to store the pages.
> * For each link you extract, check if the target is in the list you generated above. If yes, keep it. If no, discard it.
> * Use a NetworkX [`DiGraph`](http://networkx.lanl.gov/reference/classes.digraph.html) to store the network.
> 

```python
%matplotlib inline
import networkx as nx
import json, os
import glob
from matplotlib import pyplot as plt

DG = nx.DiGraph()
for phy in comb_list:
    DG.add_node(phy)
    



path_to_json = 'philosophers/'
json_files = [pos_json for pos_json in os.listdir(path_to_json)] #read all json files in that dir


for json_file in json_files:
    json_page=json.loads(open(path_to_json+json_file).read())    
    page_key=json_page['query']['pages'].keys()[0]
    page_name=json_page['query']['pages'][page_key]['title'] #phy name 
    if 'revisions' in json_page['query']['pages'][page_key].keys():
        page_info=json_page['query']['pages'][page_key]['revisions'][0]['*']
        page_links=re.findall(r'\[\[([\w\s\.-]+)([\w\s\.|()-,]*)\]\]', page_info,re.UNICODE)
        for page_link in page_links:   
            link_tag=page_link[0]
            if(link_tag in comb_list):
                DG.add_edge(page_name,link_tag)
    #if link_tag in comb_list

#print (DG.edges())
pos=nx.spring_layout(DG)
nx.draw(DG,pos,node_size=2,edge_color='r',node_color='m',width=0.5)
plt.figure(num=None, figsize=(8, 8), dpi=1200)
plt.show()
#    print (page_json)
       
    
    
```
