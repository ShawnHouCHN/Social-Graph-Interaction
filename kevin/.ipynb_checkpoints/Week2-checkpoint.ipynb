The values represented in the matrix can also be numbers of degrees in a directed weighted graph instead of just 0 and 1. That can not just be infered from the list because it only tells there is a link but not to what degree. This is illustrated in below image.

```python
from IPython.display import Image
Image(filename='img/Weighted-Graph-Adjacency-Matrix.png')
```

Any other operations can be done in both representations see this [dzone blog entry](https://dzone.com/articles/algorithm-week-graphs-and). Depending on the representation these operations can have different complexities. In case that we're using adjacency matrix we have:

* Adding an edge - $O(1);$
* Deleting an edge - $O(1);$
* Answering the question "is there an edge between i and j" - $O(1);$
* Finding the successors of a given vertex - $O(n);$
* Finding (if exists) a path between two vertices - $O(nÂ²);$

While for an adjacency list we can have:

* Adding an edge - $O(log(n));$
* Deleting an edge - $O(log(n));$
* Answering the question "is there an edge between i and j" - $O(log(n));$
* Finding the successors of a given vertex - $O(k)$, where "k" is the length of the lists containing the successors of i;
* Finding (if exists) a path between two vertices - $O(n+m$) - where $m <= n;$

We now see that depending of the representation of the graph we can have different complexities for the same operations. This is very important while trying to solve a problem and can be crucial while chosing the algorithm.

```python
print "f) how many paths of length 3 starting at 1 and ending at 3:\n"
total = 0
for idx, path in enumerate(nx.all_simple_paths(GA, source=1, target=3)):
    total = idx+1
    if len(path) == 4:
        print "3 edges between 4 nodes at:", path

print "The shortest path is:", [p for p in nx.all_shortest_paths(GA,source=1,target=3)]
```

When you also consider repetition of edges then the `NetworkX` library can not find it. Consider the shortest path `[1, 3]`, if that part is repeated `[1, 3, 1, 3]` then the conditions are also met.

```python
print "g) number of cycles == 4:\n",

# transform the undirected into directed graph
# replaces each undirected edge with two directed edges
HA = GA.to_directed()

count = 0
for x in list(nx.simple_cycles(HA)):
    print "cycle:", x
    if len(x) == 4:
        count += 1
print 'counted a total of {0} cycle(s) (of length 4)'.format(count)
```

Consider the bipartite network of Figure 2.21

```python
from IPython.display import Image
Image(filename='img/NS2.5.png')
```

```python
# a)
NetGA = nx.Graph(
    [
        (1,7),
        (2,9),
        (3,7),
        (3,8),
        (3,9),
        (4,9),
        (4,10),
        (5,9),
        (5,11),
        (6,11)
    ]
)
AM = nx.adjacency_matrix(NetGA)
print AM.todense()
```

It is a block-diagonal matrix because the blocks off the diagonals are zero matrices, and the diagonal matrices are square.

```python
# b)
purpleG = nx.Graph(
    [
        (1,3),
        (2,3),
        (2,4),
        (2,5),
        (3,4),
        (3,5),
        (4,5),
        (5,6)
    ]
)

greenG = nx.Graph(
    [
        (7,8),
        (7,9),
        (8,9),
        (9,10),
        (9,11),
    ]
)

AMpurple = nx.adjacency_matrix(purpleG)
print "adjacency matrix purple nodes:"
print AMpurple.todense(), "\n"

AMgreen = nx.adjacency_matrix(greenG)
print "adjacency matrix green nodes:"
print AMgreen.todense()
```

```python
# c)
from __future__ import division

def get_average_degree(graph, nodes=None, print_edges=False):
    if not nodes:
        nodes = graph.nodes()
    deg_graph = graph.degree(nbunch=nodes)
    sum_of_edges = sum(deg_graph.values())
    deg = sum_of_edges/len(nodes)
    if print_edges:
        return deg, deg_graph.values()
    return deg

deg_p, edg_p = get_average_degree(NetGA, [1,2,3,4,5,6], print_edges=True)
print "list of degree of edges - purple nodes:", edg_p
print "average degree bipartite network - purple nodes:", deg_p

deg_g, edg_g = get_average_degree(NetGA, [7,8,9,10,11], print_edges=True)
print "\nlist of degree of edges - green nodes:", edg_g
print "average degree bipartite network - green nodes:", deg_g
```

```python
# d)
deg_pg, edg_pg = get_average_degree(purpleG, print_edges=True)
print "list of degree of edges network projection - purple nodes:", edg_pg
print "average degree network projection - purple nodes:", deg_pg

deg_gg, edg_gg = get_average_degree(greenG, print_edges=True)
print "\nlist of degree of edges network projection - green nodes:", edg_gg
print "average degree network projection - green nodes:", deg_gg
```

It is not surprising that the values in _d)_ are different from those obtained in point _c)_ because we look at two different separated networks in _d)_. In _c)_ we just look at different nodes in one network. The purple projection network also has more edges than the respective bipartie network which influences the sum for outgoing and incoming degrees.

> * Binning real numbers
>   * Let's do a gentle start and use the `random` library generate 5000 data points from a Gaussian distribution with $\mu = 2$ and $\sigma = 0.125$.
>   * Now, let's use `numpy.histogram` to bin those number into 10 bins. What does the `numpy.histogram` function return? Do the two arrays have the same length?
>   * Then we use `matplotlib.pyplot.plot` to plot the binned data. You will have to deal with the fact that the counts- and bin-arrays have different lengths. Explain how you deal with this problem and why.


```python
# Binning real numbers

# importing the right packages
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import random as rand
import networkx as nx

# generate data points
# https://docs.python.org/2/library/random.html
mu = 2
sigma = 0.125
dat = [rand.gauss(mu, sigma) for i in xrange(5000)]
# or: dat = np.random.normal(mu, sigma, 5000)

# bin numbers
hist, bin_edges = np.histogram(dat, bins=10)
print "hist length:", len(hist), "bin_edges length:", len(bin_edges)
```

`numpy.histogram` returns two arrays, `hist` are the values of the histogram and `bin_edges` a float array with the edges of the 10 bins that has a length of`len(hist)+1`.

```python
plt.hist(dat)
plt.title("Histogram of random numbers with Gaussian distribution")
plt.xlabel("automatic bins")
plt.ylabel("amount")
plt.show()

# plot binned data
plt.bar(bin_edges[:-1], hist, width=bin_edges[1]-bin_edges[0], color='red')
plt.xlabel("bins")
plt.ylabel("amount")
plt.title("Histogram of binned data")
plt.show()
```

To get around the differently sized arrays the last edge `bin_edges[:-1]` is simply ignored and given to the first parameter in the `plt.bar` function. The additional edge comes to have a centered distribution but with just 10 bins it is not necessary for our data as the first diagram shows.

> * Binning integers
>   * But binning real numbers into a fixed number of bins is easy when `numpy.histogram` does all the work and finds the right bin boundaries for you. 
> Now we'll generate a bunch of integers and set the bin boundaries manually. This time, let's grab data from a Poisson distribution. As it turns out 
> `numpy` also has some convenient random number generators. Use `numpy.random.poission` to generate 5000 numbers drawn from a Poisson distribution 
> characterized by $\lambda = 10$. Find the maximum and minimum value of your 5000 random numbers. 
>   * Instead of simplify specifying the number of bins for `numpy.histogram`, let's specify the bins we want using a vector. Create a vector 
> that results in a binning that puts each integer value in its own bin and where the first bin contains the minimum number you found above, 
> and the last bin contains the maximum number. What is the sum over bin counts? Explain how the binning-vectors first and last element 
> relates to the min and max from the Poisson distribution.
>   * Now, use a bar chart (`matplotlib.pyplot.bar`) to plot the distribution 
 

```python
# Poisson distribution
lam = 10
dat = np.random.poisson(lam, 5000)
min_bin = min(dat)
max_bin = max(dat)
print "max val:", max_bin, "min val:", min_bin
```

```python
# list/vector from min to max
bin_range = xrange(min_bin, max_bin + 1)

# http://docs.scipy.org/doc/numpy/reference/generated/numpy.digitize.html
# indicies of what data element belongs in what bin
inds = np.digitize(dat, bin_range)

# empty list with bin_range number of elements
vec_bins = [[] for i in xrange(len(bin_range))]

# put the data element into the respective bin
for idx, ele in enumerate(dat):
    vec_bins[inds[idx]-1].append(ele)

bin_counts = [len(x) for x in vec_bins]
print "elements in bin counts:", bin_counts
print "sum over bin counts =", sum(bin_counts)
print "first and last element of binning vector:", vec_bins[0], "and", vec_bins[-1]
```

Since we created a vector in the range of the min and max of the Poisson distribution the binning-vectores first element contain the minimum values of the distribution and the maximums are contained in the last bin.

```python
# plot bar chart
plt.bar(bin_range, bin_counts, width=bin_range[1]-bin_range[0])
plt.xlabel("bins")
plt.ylabel("amount")
plt.title("Histogram of binned data")
plt.show()
```

> 

[Sune's Network](https://raw.githubusercontent.com/suneman/socialgraphs2016/master/files/edge_colormap.png)

```python
import random as rand

N = 200
k = 1.5

p = k/(N-1)
ERG = nx.erdos_renyi_graph(N, p)

# extract giant connected component, needs longest subgraph thus sorting all subgraphs after length
sub_graphs =  sorted(nx.connected_component_subgraphs(ERG), key=len)
# sub_graphs is generator of graphs, one for each connected component of ERG
gcc = sub_graphs[-1]
# choose random node
rand_node = rand.choice(gcc.nodes())
# find all nodes 2 steps away from rand_node
nodes_2steps = nx.single_source_shortest_path_length(gcc, source=rand_node, cutoff=2)
# remove the values with length 1
nodes_2steps = {key: value for key, value in nodes_2steps.items() 
             if value is not 1}

print 'Chosen random starting node is {0}\nHas following nodes => steps (key => value): {1}'.format(
    rand_node, nodes_2steps)

# plot the GCC, must be in order most to least amount of nodes
pos = nx.spring_layout(ERG, pos=nodes_2steps)
# other blue nodes
nx.draw(gcc, pos, with_labels=False, node_size=60, node_color="#A0CBE2")
# nodes 2 steps away red
nx.draw_networkx_nodes(gcc, pos,  with_labels=False, node_size=60, nodelist=nodes_2steps.keys(), node_color="#FF0000")
# starting node black
nx.draw_networkx_nodes(gcc, pos, with_labels=False, node_size=60, nodelist=[rand_node], node_color="#000000")

plt.draw()
```
